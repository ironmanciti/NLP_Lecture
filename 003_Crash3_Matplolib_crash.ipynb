## Functional Programming Style

````python
import numpy as np
from matplotlib import pyplot as plt
````

### 점찍기

````python
plt.plot(3, 2, 'o', c='r')
````

### 선긋기

````python
plt.plot([1,2,3],[4,5,1])

plt.show()
````

### title, label 추가

````python
x = [5,8,10]
y = [12,16,6]
x2 = [6, 9, 11]
y2 = [6, 15, 7]

plt.plot(x,y)
plt.plot(x2,y2)

plt.title('Epic Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
````

### bar chart

````python
plt.bar(x, y)
plt.bar(x2, y2)

plt.title('bar chart')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
````

### scatter plot (산점도)

````python
plt.scatter(x, y, label='x')
plt.scatter(x2, y2, label='x2')

plt.title('Scatter plot')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.legend()
````

## histogram

````python
x = np.random.randn(1000)

plt.hist(x, bins=10)

plt.title('histogram')
plt.ylabel('Y axis')
plt.xlabel('X axis')
````

## multiple subplots

**y = x + noise 인 (x, y) 쌍의 toy data 생성**

````python
x = np.random.randint(low=1, high=11, size=50)
y = x + np.random.randint(1, 5, size=x.size)

data = x + y
data[:5]
````

````python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

ax1.scatter(x, y, marker='o', c='r', edgecolor='b')
ax1.set_title("Scatter: x vs. y")
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.hist(data, bins=10, label=('x', 'y'))
ax2.set_title("Frequency")
ax2.legend()
ax2.yaxis.tick_right()
````
