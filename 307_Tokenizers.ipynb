## Keras 기본 Tokenizer - rule-based
- 공백 또는 구둣점으로 분리  
- 영어는 단어별로 띄어쓰기가 철저히 지켜지는 언어이므로 가능

```python
sentences_E = [
    'I love my dog',
    'I love my cat',
    'You love my dog!',
    'I was born in Korea and graduaged University in USA.',
]

sentences_K = [
    "코로나가 심하다",
    "코비드-19가 심하다",
    '아버지가방에들어가신다',
    '아버지가 방에 들어가신다',
    '너무너무너무는 나카무라세이코가 불러 크게 히트한 노래입니다'
]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(num_words=100, oov_token='<OOV>')  # 빈도수 상위 100 개로 구성

tokenizer.fit_on_texts(sentences_E)
word_index = tokenizer.word_index
print(word_index)
```

- Keras의 rule base tokenizer로 한글을 tokenize

```python
tokenizer = Tokenizer(num_words=100, oov_token='<OOV>')  # 빈도수 상위 100 개로 구성

tokenizer.fit_on_texts(sentences_K)
vocabulary_keras_korean = tokenizer.word_index
print(vocabulary_keras_korean)
```

###. 단어 사전 기반 한국어 tokenizer 사용
- Okt는 한국어 자연어 처리를 위한 Python 라이브러리인 Konlpy의 일부로 제공되는 형태소 분석기 (구 twitter)

```python
from konlpy.tag import Okt

okt = Okt()

temp_X = []
for sent in sentences_K:
    temp_X.append(okt.morphs(sent))
    print(okt.morphs(sent))
```

- 사전 기반 tokenize 후 Keras tokenizer 로 vocabulary 생성

```python
tokenizer = Tokenizer(num_words=100, oov_token='<OOV>')  # 빈도수 상위 100 개로 구성

tokenizer.fit_on_texts(temp_X)
vocabulary_okt_keras = tokenizer.word_index
print(vocabulary_okt_keras)
```

- 두 vocabulary 의 차이 비교

```python
print(vocabulary_keras_korean)
print(vocabulary_okt_keras)
```

- 단, Okt 사전에 미등록된 단어의 경우 정확한 tokenizing 이 안된다.  

okt.pos는 Okt 형태소 분석기에서 제공하는 메서드 중 하나로, 주어진 텍스트를 형태소로 분석하고 해당 형태소의 품사를 태깅해주는 기능을 수행합니다. pos는 "part-of-speech"의 약자로, 품사를 의미합니다. okt.pos 메서드를 사용하면 한국어 텍스트의 각 형태소에 대한 품사 정보를 함께 얻을 수 있습니다.

```python
okt.pos('너무너무너무는 나카무라세이코가 불러 크게 히트한 노래입니다')
```

## Subword-based Tokenization
- BERT Tokenizer

```python
from transformers import BertTokenizer

bert_tokenizer = BertTokenizer.from_pretrained("bert-base-cased")
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
```

```python
bert_tokenizer
```

```python
tokenizer
```

```python
# 사전의 크기를 확인합니다.
vocab_size = len(tokenizer.vocab)
print(vocab_size)
print()
# 사전의 일부 단어를 확인합니다.
for i, (token, index) in enumerate(tokenizer.vocab.items()):
    if i > 30:
        break
    print(token, index)
```

- BERT Tokenizer를 이용한 tokenization

```python
tokenized = tokenizer("Using a Transformer network is simple")
tokenized
```

- Subword 확인

```python
index_to_token = {i: token for token, i in tokenizer.vocab.items()}

for idx in tokenized.input_ids:
    print(index_to_token[idx], end=' ')
```

- Tokenizer의 decode를 이용한 문장 복원

```python
tokenizer.decode(tokenized.input_ids)
```

## AutoTokenizer

- tokenize  
- convert_token_to_ids  
- decode

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

sequence = "Using a Transformer network is simple"
tokens = tokenizer.tokenize(sequence)

print(tokens)
```

```python
ids = tokenizer.convert_tokens_to_ids(tokens)

print(ids)
```

```python
decoded_string = tokenizer.decode([7993, 170, 11303, 1200, 2443, 1110, 3014])
print(decoded_string)
```
