### Pre-trained model load

이미 학습된 Transformer 모델을 로드하는 것은 간단합니다. 다음 from_pretrained()방법을 사용하여 이를 수행할 수 있습니다.  

- `bert-base-cased` - [BERT 작성자가 직접 교육한 모델 체크포인트](https://huggingface.co/bert-base-cased). 훈련된 작업에 대한 추론을 위해 직접 사용할 수 있으며, 새로운 작업에 대해 미세 조정할 수도 있습니다.

```python
from transformers import BertModel    # Bert model 의 checkpoint load

bert_model = BertModel.from_pretrained("bert-base-cased")  #BERT 작성자가 직접 교육한 모델 체크포인트
print(type(bert_model))
```

```python
from transformers import AutoModel    # 범용적 checkpoint load

auto_model = AutoModel.from_pretrained("bert-base-cased")  #BERT 작성자가 직접 교육한 모델 체크포인트
print(type(auto_model))
```

```python
auto_model = AutoModel.from_pretrained("facebook/bart-base")  #BERT 작성자가 직접 교육한 모델 체크포인트
print(type(auto_model))
```

### BERT 모델을 초기화하기 위해 구성 객체 로드

처음부터 모델을 훈련하면  오랜 시간과 많은 데이터가 필요합니다.

```python
from transformers import BertConfig, BertModel

# config 구성
config = BertConfig()
# config로부터 모델 구성
model = BertModel(config)
```

- 모델을 빌드하는 데 사용되는 속성들   

```python
print(config)
```

```python
print(model)
```

Default 값인 12 개 layer 대신 10 개 layer 만 사용할 경우의 구성 변경

```python
# config 구성 변경
config = BertConfig.from_pretrained('bert-base-cased', num_hidden_layers=10)
model = BertModel(config)
model
```
